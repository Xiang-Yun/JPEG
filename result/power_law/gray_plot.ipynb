### 原始

```python
import matplotlib.pyplot as plt
%matplotlib inline

f = open("Y_graylevel.txt", 'r')
lines = f.readlines()
data = []

for line in lines:
    data.append(int(line.split(',')[0]))

plt.bar(range(len(data)), data)
plt.title('lena_GrayLevel Original')
plt.xlabel('Gray Level')
plt.ylabel('Number of pixels')
plt.show()
```
